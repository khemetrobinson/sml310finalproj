```python
import pandas as pd

df = pd.read_csv("/Users/khemetrobinson/Desktop/Spring 2021/Spring Independent Work/billboard-hot-100-yearly.csv")

song_list = df["TITLE"].tolist()
artist_list = df["ARTIST"].tolist()
```

```python
def get_lyrics(songs, artists):
    i = 0
    lyrics = []
    for song in songs:
        try:
            lyrics.append(re.sub("\n", " ", genius.search_song(songs[i], artists[i]).lyrics))
            # lyrics.append(genius.search_song(songs[i], artists[i]).lyrics)
        except:
            lyrics.append("")
            
        i += 1
    return lyrics
```

```python
lyrics_list = get_lyrics(song_list, artist_list)
```

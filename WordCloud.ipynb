```python
import pandas as pd

df = pd.read_csv('song_data.csv', encoding='utf-8') 
lyrics = df['LYRICS'].tolist()
genres = df['GENRE'].tolist()
```

```python
from wordcloud import WordCloud

i=0
all_words = ''
for i in range(0, len(df)):
  genre = genres[i]
  if genre == 'Rap':
    lyric = lyrics[i].replace('got', '')
    lyric = lyric.replace('im', '')
    lyric = lyric.replace('yeah', '')
    lyric = lyric.replace('know', '')
    all_words += lyric

word_cloud = WordCloud(width=1000, height=562, background_color='white').generate(all_words)
word_cloud.to_file('wordcloud-hiphoprap.png')
image = word_cloud.to_image()
```
